## **Data Preparation**

```python
# Create Dataframe 
basket= pd.pivot_table(data=trx_detail, index='transaction_id', columns='item_name',
                       values='item_id', aggfunc=pd.Series.nunique, fill_value=0)
basket
```

```python
# Check what menus ordered based on trnsacation_id
ro= basket.loc['ffff971e-ca8a-4863-85fd-b0deab016ccd']
nonzero= ro[ro>0].index.tolist()

basket.loc[['ffff971e-ca8a-4863-85fd-b0deab016ccd'],nonzero]
```

```python
# Check item detail ordered based on transaction ID
trx_detail[trx_detail['transaction_id']=='ffff971e-ca8a-4863-85fd-b0deab016ccd']
```

```python
# Encode Dataset
def encode(x):
    if x==0:
        return False
    if x>0:
        return True
basket= basket.map(encode)
basket
```

```python
# Take transaction that have two items or more
basket_filter= basket[(basket>0).sum(axis=1)>1]
basket_filter
```

```python
# Check

basket_filter.loc[['0002b217-3d63-4fad-9e39-c0b07e015918']]
```

## **Analysis**

```python
# USe apriori algorithm for analysis
from mlxtend.frequent_patterns import apriori

freq_item= apriori(basket_filter, min_support=.01, use_colnames=True).\
            sort_values('support', ascending=False).reset_index(drop=True)
freq_item['product_count']= freq_item['itemsets'].apply(lambda x: len(x))
freq_item
```

```python
# Summary variabels

from mlxtend.frequent_patterns import association_rules

product_assc= association_rules(freq_item, metric='confidence', min_threshold=.21).\
                sort_values(['support','confidence'],ascending=[False,False]).reset_index(drop=True)
product_assc
```

```python
product_assc.sort_values('antecedent support', ascending=False)
```

Espresso has ordered 5.9% percentage from total transaction. After buy a the Espresso, cust will ordered latte too with 21% probability. Espresso and latte have negative correlation(n<1), that means the probability of these product bought together lower than these product bought for itself.

FOr this case, there are no one combination that realy have big correlation. That can be indicated by the highest confidence value is 0.21 which means there are no combination have probability higher thatn 21% to bought them together. The most cust seems to order just one menu for each transaction.
